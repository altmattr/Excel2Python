```python
import pandas as pd

stats = pd.read_csv("data/afl/stats.csv")
games = pd.read_csv("data/afl/games.csv")
players = pd.read_csv("data/afl/players.csv")

full_stats = stats.merge(games, left_on="gameId", right_on="gameId").merge(players, left_on="playerId", right_on="playerId")
```

```python
full_stats.groupby("venue").max().sort_values("Disposals", ascending=False)
full_stats.info()
```

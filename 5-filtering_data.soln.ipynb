# Solution - regex

```python
import pandas as pd

a_grade = pd.read_csv("data/wsmtb/2018-XC-Club-Champs.csv", skiprows = [0,1,3], nrows=8)
times = a_grade[["Race Time", "Lap1", "Lap2", "Lap3", "Lap4", "Lap5"]]

times["Race Time"].replace(",\d", "", regex=True)
```

```python
times["Race Time"].replace(",\d", ".0000", regex=True)
```

```python
times.replace("(.*),(\d)", "\\1.\\2", regex=True)
```

# Solution - lithgow filter

We show two solutions, one that takes two steps but only uses what we know.  This solution creates a data frame with just data after 1990 and then filters that one to have only data before 2000.  The second solution uses a more sophisticated "mask" but the syntax is harder to get right.  I recommend you experiment with the syntax in the second version to see what you can make work and what doesn't work.

```python
lithgow = pd.read_csv("data/rainfall/lithgow.csv")

after_1990 = lithgow[lithgow["Year"] >= 1990]
after_1990_and_before_2000 = after_1990[after_1990["Year"] < 2000]
after_1990_and_before_2000
```

# Solution - regex

```python
import pandas as pd

a_grade = pd.read_csv("data/wsmtb/2018-XC-Club-Champs.csv", skiprows = [0,1,3], nrows=8)
times = a_grade[["Race Time", "Lap1", "Lap2", "Lap3", "Lap4", "Lap5"]]

times["Race Time"].replace(",\d", "", regex=True)
```

```python
times["Race Time"].replace(",\d", ".0000", regex=True)
```

```python
times.replace("(.*),(\d)", "\\1.\\2", regex=True)
```

At this point you can see that the indexes are not working for us since they are not unique.  Perhaps we should have imported that data some other way?  `read_csv` lets us take just the bits we need.  `skiprows` and `nrows` are what we want

```python
a_grade = pd.read_csv("data/wsmtb/2018-XC-Club-Champs.csv", skiprows = 3, nrows=8)
a_grade
```

It looks like row 2 has the header data we want, we can used a more advanced form of `skiprows` leave that header row at the top where we need it

```python
a_grade = pd.read_csv("data/wsmtb/2018-XC-Club-Champs.csv", skiprows = [0,1,3], nrows=8)
a_grade
```

The data here seems to be time data - pandas has a nice type `TimeDelta` that we could use for that.  Also, `pandas` can convert columns for us into this data type

```python
pd.to_timedelta(a_grade["Race Time"])
```

Its not quite right, I need to strip the centi-seconds from the time.  I do this with a regex - much better.

```python
pd.to_timedelta(a_grade["Race Time"].str.replace(",\d", "", regex=True))
```

```python
for col in ["Race Time", "Lap1", "Lap2", "Lap3", "Lap4", "Lap5", "Lap6"]:
  a_grade[col] = pd.to_timedelta(a_grade[col])
a_grade
```

Lets add some cummulative time columns

```python
for col in [1,2,3,4,5,6]:
    so_far = pd.Timedelta(0)
    for c in range(1,col+1):
        so_far = so_far + a_grade[f"Lap{c}"]
    a_grade[f"Cumm{col}"] = so_far 
a_grade
```

```python
a_grade[["Lap1", "Lap2", "Lap3", "Lap4", "Lap5", "Lap6"]].transpose().plot()
```

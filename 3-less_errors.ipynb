```python
import pandas as pd

whr = pd.read_excel("data/whr.xlsx", "2017")
whr.info()
```

```python

happiness_threshold = 7
family_threshold = 1.5
whr["Happiness?"] = whr["Happiness.Score"] > happiness_threshold
whr["Family but not happy"] = (whr["Family"] > family_threshold) & (whr["Happiness?"] == False)
whr
```

```python
# who are these countries that have strong family values but are not very happy?

whr[whr["Family but not happy"]] # we will see how this trick works next week :)
```

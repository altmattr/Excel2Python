```python
import pandas as pd

agrade = pd.read_csv("2022-XC-Round-3-AGrade.csv", index_col="name")
bgrade = pd.read_csv("2022-XC-Round-3-BGrade.csv", index_col="name")
cgrade = pd.read_csv("2022-XC-Round-3-CGrade.csv", index_col="name")
dgrade = pd.read_csv("2022-XC-Round-3-DGrade.csv", index_col="name")
race = pd.concat([agrade, bgrade, cgrade])
race
```

```python
race["total1"] = pd.to_timedelta(race["lap1"])
race["total2"] = race["total1"] + pd.to_timedelta(race["lap2"], errors="coerce")
race["total3"] = race["total2"] + pd.to_timedelta(race["lap3"], errors="coerce")
race["total4"] = race["total3"] + pd.to_timedelta(race["lap4"], errors="coerce")
race["total5"] = race["total4"] + pd.to_timedelta(race["lap5"], errors="coerce")

race[["total1","total2","total3","total4", "total5"]]
```

```python
import numpy as np
# add gap to leader
race["gap0"] = 0.0
for lap in [1,2,3,4,5]:
    min = race[f"total{lap}"].min()
    print(min)
    print(not pd.isnull(min))
    if (not pd.isnull(min)):
      race[f"gap{lap}"] = (race[f"total{lap}"] - race[f"total{lap}"].min())/race[f"total{lap}"].min()
    else:
      race[f"gap{lap}"] = np.NaN
race[["gap0","gap1", "gap2", "gap3", "gap4", "gap5"]]
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 10]

race[["gap0","gap1", "gap2", "gap3", "gap4", "gap5"]].transpose().plot(ylim=(0,0.5), legend=False)
```

```python
race["ll"] = pd.to_datetime(race["lastlap"])
race["ml"] = pd.to_datetime(race["minlap"])
race.plot(kind="scatter", x="ll", y="ml")
```

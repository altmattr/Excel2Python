```python
import pandas as pd
```

```python
stats = pd.read_csv("data/afl/stats.csv")
games = pd.read_csv("data/afl/games.csv")
players = pd.read_csv("data/afl/players.csv")
```

```python
bags = stats["Goals"] >= 6
stats[bags]
```

# Solution - Ruckman kicks a bag?

Can you find any players who played as a "Ruck" (you will find this in the "position" column of the `players` file) and who kicked a bag?

```python
bag_players = stats[bags].merge(how="left", left_on="playerId", right=players, right_on="playerId")
rucks = bag_players["position"].str.contains("Ruck")
bag_players[rucks]
```
